```python
#df.groupby('A').agg({'B': ['min', 'max'], 'C': 'sum'})

def multiply(income,weight,tar):
    i = 0
    while i < weight:
        tar.append(income)
        i+=1

def explode_weights(df,weight_col):
    ## Make a list of HHINCOME for calculating
    ## a quantile where each household's HHINCOME
    ## is repeated in the series n times, where
    ## n == weight_col
    tar = []
    df.apply(lambda x: multiply(x["HHINCOME"],x[weight_col],tar),axis=1)
    return tar


## There are a couple ways to use this data to
## estimate the 95th percentile of income. 1 is on
## the household level and the other is at the person
## level. It seems strange to combine HHINCOME and PERWT,
## but this also seems to yield the most sensible results...
## My numbers still don't align with Brookings...I could 
## request another data extract with personal income
##
def simulate_data(msa_year,hh):
    weight_col = "HHWT"
    if hh:
        ## households -- compress the data to the level of households
        ## then explode out the weights (HHWT)
        data = msa_year[["SERIAL","MET2013","HHINCOME","HHWT"]].groupby("SERIAL").agg("max")
    else:
        ## leave the data so that 1 row corresponds to 1 person, then
        ## explode out the weights (PERWT) <-- This seems closest to 
        ## Brookings' numbers?
        data = msa_year[["SERIAL","MET2013","HHINCOME","PERWT"]]
        weight_col="PERWT"
    return explode_weights(data,weight_col)
    
def weighted_quantile(msa_year,quant,hh=True):
    simulated_dist = pandas.Series(simulate_data(msa_year,hh))
    return simulated_dist.quantile(quant)

weighted_quantile(nyc_2014,.95,False)
```

```python
top_coded = acs_14[acs_14["HD01_VD06"].isnull()]
pums_14 = pums[pums["YEAR"]==2014]

def city_df(geo_id,pums):
    return pums[pums["MET2013"]==geo_id]

top_coded["PUMS_Est"] = top_coded["GEO.id2"].apply(lambda x: weighted_quantile(city_df(x,pums_14),.95,False))
top_coded["PUMS_Est_20"] = top_coded["GEO.id2"].apply(lambda x: weighted_quantile(city_df(x,pums_14),.2,False))
```

Track either mean wages per quintile or quintile cut-offs over time, to show the stratification of the recession and the recovery. Probably want to do this for something like:

So you need a call for each year -- and the url construction might change year to year, though I think it's mostly just 2016 that's special for some reason.

A bunch of maybe-useful api queries:

    * https://api.census.gov/data/2016/acs/acs1/variables.html
    * https://api.census.gov/data/2016/acs/acs1/examples.html
    * https://api.census.gov/data/2016/acs/acs1?get=B01001_001E&for=metropolitan%20statistical%20area/micropolitan%20statistical%20area:*
    * https://api.census.gov/data/2015/acs1/examples.html
    * https://api.census.gov/data/2016/
            

```python
import json,urllib,re,numpy,time,config

base = "https://api.census.gov/data/"

def construct_url(year,geo_type,geo_code):
    url = base+str(year)+"/acs/acs1?get="+config.fec_key+"&is_amended=False&per_page=100"
    return(url)

def lookup(cid, t=5):
    url = construct_url(cid)
    data = url_to_data(url)
    if t>0:
        time.sleep(t)
    if data is None:
        return None
    return data["results"]

## If the fn can't find a response at the url, it returns None
## That gets converted into an empty dataframe (not ideal), and
## and flagged for the second pass
def url_to_data(url):
    try:
        req = urllib.request.Request(url)
        req.add_header('User-agent', 'Mozilla 5.10')
        res = urllib.request.urlopen(req)
        data = json.load(res)
        return data
    except urllib.error.HTTPError as err:
        return None
    
lookup("C00406389")
```
